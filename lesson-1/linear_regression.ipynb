```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('income.csv')
df=df.rename(columns={'per capita income (US$)':'income'})
df.head(5)
```

```python
%matplotlib inline
plt.xlabel('income')
plt.ylabel('year')
plt.scatter(df.income,df.year,color='red',marker='_')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']].values.reshape(-1,1),df.income)
predict_income = reg.predict([[1980]])
predict_income
```

```python

plt.scatter(1980,predict_income,color='yellow',marker='o')
plt.scatter(df.year,df.income,color='red',marker='_')
plt.show()
```

```python
plt.scatter(df.year,df.income,color='red',marker='_')
plt.plot(df.year,reg.predict([[1980]]).repeat(47,axis=0),color='blue')
plt.plot(df[['year']].values.reshape(-1,1),df.income,color='green')
plt.show()
```

```python
slope = reg.coef_[0]
slope
```

```python
intercept = reg.intercept_
intercept
```

```python
plt.scatter(df.year,df.income,color='red',marker='_')
plt.plot(df.year,slope*df.year+intercept,color='green',marker='_')
plt.show()
```
