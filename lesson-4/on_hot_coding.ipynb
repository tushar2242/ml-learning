```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('carprices.csv')
df
```

```python
df =df.rename({'Sell Price($)':'SellPrice','Age(yrs)':'Age' },axis='columns')
df
```

```python
dummy = pd.get_dummies(df['Car Model'])
merge = pd.concat([df,dummy],axis='columns')
merge
```

```python
df = merge.drop(['Car Model','Mercedez Benz C class'],axis='columns')
df
```

```python
X=df.drop('SellPrice',axis='columns')
X
```

```python
y=np.array(df.SellPrice)

reg =LinearRegression()
```

```python
reg.fit(X,y)
```

```python
reg.score(X,y)
```

```python
reg.predict([[45000,4,0,0]])
```
